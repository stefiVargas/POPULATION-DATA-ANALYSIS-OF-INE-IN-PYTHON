```python
## vamos a construir una función que nos pinte el gráfico con el fin de no estar copiando y pegando la función anterior
## para cada serie : hombre, mujer, total. La serie toma una serie de parametros y nos devuelve un gráfico tipo plob
```

```python
def plot_df(x,y,title, xlabel, ylabel):    
    plt.figure(figsize = (9,4))
    plt.plot(x,y, color = 'tab:red');
    plt.xlabel (xlabel)
    plt.ylabel (ylabel)
    plt.title (title)
    plt.xticks (rotation = 45)
    plt.grid()
    plt.show() 
```

```python
df.dato_2.unique() #nos devuelve la seria de datos, como es un array se puede recorrer con un bucle for, como una lista
## para poder llamar a la función anterior plot_df, debemos recorrerla con un bucle for 
```

```python
for i in df.dato_2.unique():
    plot_df(df.fecha[df.dato_2 == i] ,df.poblacion[df.dato_2 == i],'Poblacion {0}'.format(i),'Fecha','Población(millones)')
```
